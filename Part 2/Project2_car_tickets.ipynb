```python
from collections import namedtuple

file = 'files/nyc_parking_tickets_extract.csv'

counter = 0

def data_generator(filename):
    with open(filename) as f:
        for line in f:
            yield line
a = []
def data_cleaner(filename):
    gen = data_generator(filename)
    column_names = next(gen).strip('\n').split(',')
    column_names = [name.replace(' ', '_') for name in column_names]
    Car = namedtuple('Car', column_names)
    for i in gen:
        car_params = next(gen).strip('\n').split(',')
        print(car_params)
        car = Car(*car_params)
        
        a.append(car)
    
data_cleaner(file)
```
