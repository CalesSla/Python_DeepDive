```python
from time import perf_counter as pc
from collections import namedtuple
```

```python
Timings = namedtuple('Timings', 'timing_1 timing_2 abs_diff rel_diff_percentage')

def compare_timigs(timing_1, timing_2):
    rel_diff_percentage = (timing_2 - timing_1) / timing_1 * 100
    abs_diff = timing_2 - timing_1
    timings = Timings(round(timing_1, 1), round(timing_2, 1), round(abs_diff, 1), round(rel_diff_percentage, 2))
    return timings

compare_timigs(1, 2)
```

```python
test_repeats = 10_000_000

import math

start = pc()

for _ in range(test_repeats):
    math.sqrt(2)

end = pc()
elapsed_fully_qualified = end - start

print(f'Elapsed time: {elapsed_fully_qualified}')   
```

```python
from math import sqrt

start = pc()

for _ in range(test_repeats):
    sqrt(2)

end = pc()
elapsed_direct_symbol = end - start

print(f'Elapsed time: {elapsed_direct_symbol}')   
```

```python
compare_timigs(elapsed_fully_qualified, elapsed_direct_symbol)
```
